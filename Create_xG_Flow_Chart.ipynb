```python
# import packages
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# read csv
df = pd.read_csv('xgtutorial.csv')

df
```

```python
# criar listas a 0 para que o chart comece no 0
a_xg = [0]
h_xg = [0]
a_min = [0]
h_min = [0]

# criar variaveis para home e away
hteam = df['team'].iloc[0] # primeira instância
ateam = df['team'].iloc[-1] # última instância

# for a correr o df inteiro e a espetar os xg e mins nas listas
for x in range(len(df['xG'])):
    if df['team'][x] == hteam:
        h_xg.append(df['xG'][x])
        h_min.append(df['minute'][x])
    if df['team'][x] == ateam:
        a_xg.append(df['xG'][x])
        a_min.append(df['minute'][x])

a_xg
```

```python
# criar função para tornar o valor cumulativo
def nums_cumulative_sum(nums_list):
    return [sum(nums_list[:i+1]) for i in range(len(nums_list))]

h_cumulative = nums_cumulative_sum(h_xg)
a_cumulative = nums_cumulative_sum(a_xg)

a_cumulative
```

```python
# plot the chart
fig, ax = plt.subplots(figsize = (10,5))
fig.set_facecolor('#3d4849')
ax.patch.set_facecolor('#3d4849')

# marcações no eixo xx
plt.xticks([0,15,30,45,60,75,90])
plt.xlabel('Minute')
plt.ylabel('xG')

ax.step(x=h_min, y=h_cumulative,where='post')
ax.step(x=a_min, y=a_cumulative,where='post')
```
